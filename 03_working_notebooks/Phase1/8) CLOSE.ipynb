# CLOSE 

This notebook will contain the first implentation of the CLOSE framework and will facilitate understanding

Fix k and the start and then run the CLOSE algorithm for stability. For this would need to take overlap of users or weight them somehow by the proportion of missing users.

The difficulty with the CLOSE metric and clustering stability metrics in general is that user - time bin do no appear every week. So we having 'missing points' and hence we can directly apply the CLOSE framework. A workaround could be to take the intersection of the weeks so that each week has the same users.

```python
# Main packages 
import polars as pl
import matplotlib.pyplot as plt
import numpy as np

# Clustering packages
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score,adjusted_rand_score

# Parallel processing packages
from joblib import Parallel, delayed
```

```python
agg_hour_level = 1
buckets_per_week = (7 * 24) // agg_hour_level  

weeks = [
    pl.read_parquet(f'/home/ma/t/tr725/MSc-Project/03_working_notebooks/results/week{w}.parquet')
    .with_columns((pl.col('bucket') % buckets_per_week).alias('bucket'))
    for w in range(1, 9)
]

def jaccard_all(dfs):
    sets = [set(zip(df['src_user'].to_list(), df['bucket'].to_list())) for df in dfs]
    intersection = sets[0].intersection(*sets[1:])
    union = sets[0].union(*sets[1:])
    return len(intersection) / len(union) if union else 0.0

print(f"Jaccard (all 8 weeks): {jaccard_all(weeks):.4f}")
```
