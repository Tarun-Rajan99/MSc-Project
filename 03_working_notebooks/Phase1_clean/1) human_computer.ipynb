```python
# Packages 
import polars as pl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Load data
df = pl.scan_csv("/home/lanl/data/cyber1/auth.txt.gz",has_header=False,separator=",",
                 new_columns= ['time','src_user','dest_user','src_comp','dest_comp',
                               'auth_type','logon_type','auth_orientation','outcome'])
```

```python
# Keep only human users
humans = df.filter(pl.col("src_user").str.starts_with("U")) 
```

```python
humans.show(5)
```

```python
# Number of human auth events
humans.select(pl.len()).collect().item()
```

```python
# Number of unique source users
humans.select(pl.col("src_user").n_unique()).collect().item()
```
